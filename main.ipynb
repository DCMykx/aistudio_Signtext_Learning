## 项目概述
手语通·SignLink是一款基于人工智能的手语知识助手，旨在为用户提供专业、易懂、有趣的手语知识学习、手语语法翻译口语以及手语课程教案编写服务，创建独属于中国手语的AI学习平台！项目整合了**《中国手语语言学》教材（446个知识条目）、中国手语词目及手势操作（6026条）、中国手语概论QA文档（100个问答对）**等权威资源，构建了全面、精准的手语知识体系。

## 核心功能
1. **中国手语知识库** 📖
   - 整合权威手语教材和词目
   - 覆盖基础词汇及手势表达、语法规则、文化差异等内容

2. **智能问答与解析** 💡
   - 支持自然语言提问
   - 自动检索知识库，生成通俗易懂的解析
   - 配以生动的手势描述和小图标

3. **手语翻译服务** ✨
   - 将指定句子翻译成手语文字表达
   - 提供详细的手势步骤、表情与身体动作描述
   - 支持趣味化表达

4. **手语课程教案编写** ✨
   - 教案主题解析与例句生成
   - 多句批量处理能力
   - 支持从教案主题自动推导教学例句
   - 优质手语课程教案生成

## 技术实现

### 数据处理流程
1. **PDF内容提取**
   - 使用pdfplumber库提取PDF文本内容
   - 使用百度OCR API识别PDF页面内容
   - 支持断点续传和结果缓存

2. **手语语言学知识、手语词目及手势操作描述提取**
   - 使用Deepseek API从文本中提取中国手语语言学知识、中国手语词目以及对应手势操作描述
   - 支持批量处理和结果去重
   - 将提取结果保存为JSON和Excel格式

3. **QA对生成**
   - 使用Deepseek模型生成中国手语概论问答对
   - 支持长文本分批处理
   - 自动修复JSON格式问题

### 主要代码模块
1. **extract_sign_text.py**
   - 提取中国手语概论PDF文本内容
   - 生成手语知识问答对
   - 保存结果到Excel

2. **extract_sign_language.py**
   - 提取手语词目和操作描述
   - 处理OCR结果
   - 保存提取结果

3. **extract_sign_knowledge.py**
   - 提取手语语言学知识
   - 处理长文本
   - 保存知识条目

## 代码说明

### extract_sign_text.py
- **功能**: 从PDF文件中提取文本内容，并生成手语概论知识问答对
- **主要函数**:
  - `extract_pdf_text(pdf_path)`: 提取PDF文本内容
  - `generate_qa_pairs(text, num_pairs=15)`: 生成问答对
  - `save_qa_to_excel(qa_pairs, output_path)`: 保存QA对到Excel文件

### extract_sign_language.py
- **功能**: 提取手语词目和操作描述
- **主要函数**:
  - `extract_text_with_baidu_ocr(pdf_path, start_page=16, end_page=None, save_path="ocr_result.txt")`: 使用百度OCR提取PDF内容
  - `extract_sign_language_entries(text, output_path="extracted_entries.json")`: 提取手语词目
  - `save_to_excel(entries, output_path)`: 保存提取结果到Excel
